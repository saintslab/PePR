```python
import pandas as pd
import numpy as np
```

```python
datasets = ['derma_pt','lidc','lidc_small','derma', \
        'derma_small','derma_smallest','pneumonia','pneumonia_small']
```

```python
df = pd.read_csv('data/full_data_pepr_dataset_level.csv')
```

```python
rDf = df[['dataset','model','type','efficient','num_param','memR','energy','train_time','test_09','pepr_e']]
```

```python
rDf.loc[:,'type'] = rDf.loc[:,'type'] == 'CNN'
```

```python
# rDf = rDf.groupby('model').mean().reset_index()
```

```python
for d in datasets:
    tmp = rDf[rDf.dataset == d]
    # tmpMax = rDf[rDf.dataset == d].sort_values(by=['best_test'],asce)
    # tmp = tmp.sort_values(by=['pepr_e'],ascending=False)
    dDf = tmp.sort_values(by=['pepr_e'],ascending=False)[:6]
    dDf.iloc[-1,:] = tmp.sort_values(by=['test_09'],ascending=False)[:1]
    dDf.loc[:,'num_param'] = dDf.num_param/1e6
    dDf.loc[:,'energy'] = dDf.energy*1000
    dDf = dDf.iloc[:,1:]
    print(d)
    print(dDf.to_latex(index=False, float_format="{:.4f}".format))
    # tmp.iloc[-1,:] = (rDf[rDf.dataset == d]).loc[df['best_test'].idxmax()]
```
